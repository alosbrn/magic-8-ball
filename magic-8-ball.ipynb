```python
import random
```

```python
# List of responses for use by the Magic 8 Ball.
# Categorization of responses is for readability only.
responses = [
    # Affirmative Responses
    "It is certain.",
    "It is decidedly so.",
    "Without a doubt.",
    "Yes definitely.",
    "You may rely on it.",
    "As I see it, yes.",
    "Most likely.",
    "Outlook good.",
    "Yes.",
    "Signs point to yes.",

    # Non-committal Responses
    "Reply hazy, try again.",
    "Ask again later.",
    "Better not tell you now.",
    "Cannot predict now.",
    "Concentrate and ask again.",

    # Negative Responses
    "Don't count on it.",
    "My reply is no.",
    "My sources say no.",
    "Outlook not so good.",
    "Very doubtful."
]
```

```python
# Prompt the user for their question.
question = input("Ask the Magic 8 Ball a yes or no question: ")
```

```python
# Sanitize the user input, removing leading/trailing whitespace.
# Print a random response, or an error if nothing is entered.
if question.strip():
    print(responses[random.randint(0, 19)])
else:
    print("Try asking a question next time!")
```
